### The original CalEnviroScreen dataset contains 68 variables related to pollution exposure, environmental conditions, health outcomes, and demographic characteristics. While additional variables such as poverty, housing burden, education, and demographic composition may be useful for exploring environmental justice impacts, the initial analysis focuses on variables most relevant to the project's objective. Pollution Score, PM2.5, Diesel PM, Traffic, Ozone, Asthma, Cardiovascular Disease, Diabetes, and Low Birth Weight were selected because they capture both environmental exposures and potential health effects associated with port activity while keeping the dataset focused and manageable.


```python
env_clean = env[
    [
        "tract",
        "county",
        "apx_loc",
        "TotalPop",
        "PollutionScore",
        "pm",
        "diesel",
        "traffic",
        "ozone",
        "asthma",
        "cvd",
        "diabetes",
        "lbw"
    ]
]
env_clean.head()
env_clean = env_clean.rename(columns={
    "tract": "census_tract",
    "apx_loc": "location",
    "TotalPop": "population",
    "pm": "pm25",
    "diesel": "diesel_pm",
    "traffic": "traffic_density",
    "cvd": "cardiovascular_disease",
    "lbw": "low_birth_weight"
})
env_clean.head()
```

```python
env_clean.isnull().sum()
env_clean = env_clean.dropna()
env_clean.shape
env_clean.info()
```

#### The CalEnviroScreen dataset was cleaned by selecting variables most relevant to the project's objective of examining the relationship between port activity, pollution exposure, and health outcomes. Variables related to overall pollution burden, specific pollutants, traffic density, and health indicators were retained, while unrelated variables were removed to improve readability and efficiency. Column names were renamed to more descriptive labels, and rows containing missing values were removed. The final cleaned dataset contains 8,859 census tracts and 13 variables, providing a consistent dataset for future analysis.


```python
env_clean.to_csv(
    "../data/cleaned/env_clean.csv",
    index=False
)
port.to_csv(
    "../data/cleaned/port_clean.csv",
    index=False
)
```
